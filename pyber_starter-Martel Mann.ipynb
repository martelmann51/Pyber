```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data

city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset

merge_table = pd.merge(city_df, ride_df, on="city")

# Display the data table for preview
merge_table
```

```python
#Grouped the Merge Data by Type & City

Rural_Type = merge_table.loc[merge_table["type"] =="Rural"]
Suburban_Type = merge_table.loc[merge_table["type"] =="Suburban"]
Urban_Type = merge_table.loc[merge_table["type"] =="Urban"]

Group_Rural = Rural_Type.groupby(["city"])
Group_Suburban = Suburban_Type.groupby(["city"])
Group_Urban = Urban_Type.groupby(["city"])

Rural_Total_Rides = Group_Rural["city"].count()
Suburban_Total_Rides = Group_Suburban["city"].count()
Urban_Total_Rides = Group_Urban["city"].count()

Rural_Average_Fare = Group_Rural["fare"].mean()
Suburban_Average_Fare = Group_Suburban["fare"].mean()
Urban_Average_Fare = Group_Urban["fare"].mean()

Total_Rural_Drivers = Group_Rural["driver_count"].mean()
Total_Suburban_Drivers = Group_Suburban["driver_count"].mean()
Total_Urban_Drivers = Group_Urban["driver_count"].mean()


```

```python
#Define X-axis and Y-axis
x_axis = Rural_Total_Rides
x_axis2 = Suburban_Total_Rides
x_axis3 = Urban_Total_Rides


y_axis = Rural_Average_Fare
y_axis2 = Suburban_Average_Fare
y_axis3 = Urban_Average_Fare

size1 = Total_Rural_Drivers*8
size2 = Total_Suburban_Drivers*8
size3 = Total_Urban_Drivers*8
```

## Bubble Plot of Ride Sharing Data

```python
#Make a Scatter Plot using Total Rides Per City and Average Fare

plt.scatter(x_axis, y_axis, marker="o", facecolors="gold", edgecolors="black",
            s=size1, alpha=1.0)

plt.scatter(x_axis2, y_axis2, marker="o", facecolors="lightblue", edgecolors="black",
            s=size2, alpha=1.0)

plt.scatter(x_axis3, y_axis3, marker="o", facecolors="tomato", edgecolors="black",
            s=size3, alpha=1.0)

plt.suptitle("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.text(45, 30, "Note: Circle size correlates with driver count per city")
plt.legend(['Rural', 'Suburban', 'Urban'])
plt.grid()

plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents

Total_Fare = merge_table["fare"].count()
Total_Rural_Fare = Rural_Type["fare"].sum()
Total_Suburban_Fare = Suburban_Type["fare"].sum()
Total_Urban_Fare = Urban_Type["fare"].sum()
Percent_Rural_Fare = (Total_Rural_Fare/Total_Fare)*100
Percent_Suburban_Fare = (Total_Suburban_Fare/Total_Fare)*100
Percent_Urban_Fare = (Total_Urban_Fare/Total_Fare)*100
# Build Pie Chart

Pie_Data = [Total_Rural_Fare],[Total_Suburban_Fare],[Total_Urban_Fare]

Pie_Data
```

```python
# Build Pie Chart

# Labels for the sections of our pie chart
labels = ["Rural", "Suburban", "Urban"]

# The values of each section of the pie chart
sizes = [4237.9299999999999, 19356.33, 39854.38000000005]

# The colors of each section of the pie chart
colors = ["lightblue", "orange", "lightcoral"]

# Tells matplotlib to seperate the "Python" section from the others
explode = (0.0, 0.0, 0.1)

# Creates the pie chart based upon the values above
# Automatically finds the percentages of each part of the pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("% of Total Fares by City Type")
```
